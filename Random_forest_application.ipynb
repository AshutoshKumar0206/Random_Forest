```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
x = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()
svc = SVC()
lr = LogisticRegression()
```

```python
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
gb.fit(x_train, y_train)
y_pred = gb.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
rf = RandomForestClassifier(max_samples=0.75, random_state=42)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(RandomForestClassifier(max_samples=0.75), x, y, cv=10, scoring="accuracy"))
```

```python
#Number of trees in random forest
n_estimators = [20, 60, 100, 120]

#Number of features to consider at every split
max_features = [0.2, 0.6, 1.0]

#Maximum number of levels in tree
max_depth = [2, 8, None]

# Number of samples
max_samples = [0.5, 0.75, 1.0]

#here multiple combinations of random forest will be trained, according to above criteria 108 combinations will be trained.
```

```python
param_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'max_samples': max_samples,
}
print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV
rf_grid = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)
```

```python
rf_grid.fit(x_train, y_train)
```

```python
rf_grid.best_params_
```

```python
rf_grid.best_score_
```

RandomSearchCV

```python
# Now when dataset is big and hyperparameters are many then grid search cv is slow so use random search cv to tune them.

#Number of trees in random forest
n_estimators = [20, 60, 100, 120]

#Number of features to consider at every split
max_features = [0.2, 0.6, 1.0]

#Maximum number of levels in tree
max_depth = [2, 8, None]

# Number of samples
max_samples = [0.5, 0.75, 1.0]

#Bootstrap samples
bootstrap = [True, False]

#Minimum number of samples required to split a node
min_samples_split = [2, 5]

#Minimum number of samples required at each leaf
min_samples_leaf = [1, 2]
```

```python
param_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'max_samples': max_samples,
    'bootstrap':bootstrap,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf,
}
print(param_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf_random = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, cv=5, verbose=2, n_jobs=-1)
#Randomized search CV randomly picks up some candidates randomly during hyperparameter tuning and leaves others
```

```python
rf_random.fit(x_train, y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```
